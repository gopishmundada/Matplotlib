# Load Required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Data

```python
gas = pd.read_csv('gas_prices.csv')
print(gas.head(5))
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

## Line Graph

```python
plt.title('Gas prices', fontdict={'fontweight':'bold','fontsize':15})
plt.plot(gas.Year, gas.USA, 'b.-')
plt.plot(gas.Year, gas.Canada, 'r+-')
plt.plot(gas.Year, gas['South Korea'], 'g.-')
#plt.figure(figsize=(20,50))
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('Years', fontdict={'fontweight':'bold','fontsize':15})
plt.ylabel('US $',  fontdict={'fontweight':'bold','fontsize':15})
plt.show()
plt.savefig('Gas.png', dpi=300)
#plt.legend(['USA'])
#plt.savefig('Gas_Priyghjkces.jpeg', dpi=300)
```

```python
plt.title('Gas prices',fontdict={'fontweight':'bold'})
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker='*')
plt.xticks(gas.Year[::3])
plt.xlabel('Years')
plt.ylabel('US $')
plt.show()
```

## Histogram Graph

```python
plt.hist(fifa.Overall,color = '#34eb98')
plt.xlabel('Skills rate')
plt.ylabel('Number Of Players')
plt.title('FIFA Details for Players', fontdict = {'fontweight':'bold','fontsize':'12'})
plt.savefig('Histogram.png', dpi=300)
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['Left','Right']
colors = ['#f22c6e','b']
explode = (0,.4)
plt.pie([left,right], labels = labels, colors = colors, autopct = '%.2f' ,explode=explode) #use of explode
plt.title('FIFA Details for Players', fontdict = {'fontweight':'bold','fontsize':'12'})
plt.show()
plt.savefig('Piechart1.png', dpi=300)

plt.pie([left,right], labels = labels, colors = colors, autopct = '%.2f')
plt.title('FIFA Details for Players', fontdict = {'fontweight':'bold','fontsize':'12'})
plt.show()
plt.savefig('Piechart.png', dpi=300)
```

### BoxPlot

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True)

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0', linewidth=2 )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
plt.savefig('Piechart.png', dpi=300)
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```
